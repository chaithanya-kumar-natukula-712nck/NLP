# NLP Tokenization using NLTK

Tokenization is the process of breaking text into smaller units called **tokens**.  
These tokens can represent **sentences, words, or phrases**, depending on the tokenizer used.

Tokenization is one of the most important preprocessing steps in **Natural Language Processing (NLP)** because most NLP algorithms cannot work directly with raw text. Converting text into tokens allows machines to analyze and process language effectively.

In this notebook, we explore several tokenizers provided by the **NLTK library** and observe how each tokenizer splits the same text differently. This helps us understand **how different tokenization strategies affect the representation of textual data** and when each tokenizer may be useful.

By comparing the outputs of different tokenizers, we can see how punctuation handling, whitespace handling, and linguistic rules influence the tokenization process.


## Tokenizers Covered in This Notebook

- Sentence Tokenizer  
- Word Tokenizer  
- WordPunct Tokenizer  
- Treebank Word Tokenizer  
- Regexp Tokenizer  
- Tweet Tokenizer  
- Toktok Tokenizer  
- Whitespace Tokenizer  
- Punkt Sentence Tokenizer  
- Multi Word Expression Tokenizer (MWETokenizer)  
- Blankline Tokenizer

```python
from nltk.tokenize import sent_tokenize,word_tokenize,wordpunct_tokenize,TreebankWordTokenizer
from nltk.tokenize import RegexpTokenizer,TweetTokenizer,MWETokenizer,ToktokTokenizer
from nltk.tokenize import BlanklineTokenizer,WhitespaceTokenizer,PunktTokenizer,PunktSentenceTokenizer
```

```python
data="Personality development encompasses the dynamic construction and deconstruction of integrative characteristics that distinguish an individual in terms of interpersonal behavioral traits. Personality development is ever-changing and subject to contextual factors and life-altering experiences. Personality development is also dimensional in description and subjective in nature. That is, personality development can be seen as a continuum varying in degrees of intensity and change. It is subjective in nature because its conceptualization is rooted in social norms of expected behavior, self-expression, and personal growth. The dominant viewpoint in personality psychology indicates that personality emerges early and continues to develop across one's lifespan. Adult personality traits are believed to have a basis in infant temperament, meaning that individual differences in disposition and behavior appear early in life, potentially before language of conscious self-representation develop. The Five Factor Model of personality maps onto the dimensions of childhood temperament. This suggests that individual differences in levels of the corresponding personality traits (neuroticism, extraversion, openness to experience, agreeableness, and conscientiousness) are present from young ages."
```

## Sentence Tokenization

Sentence tokenization splits a paragraph into individual sentences.


```python
documents=sent_tokenize(data)
print("No of Sentences in Corpus:",len(documents))
print("Sentences in Corpus:")
for doc in documents:
    print(doc)
```


In this notebook we use the `sent_tokenize()` function from NLTK, which internally uses the Punkt sentence tokenizer model. This model detects sentence boundaries using punctuation and statistical patterns in language.

When we apply `sent_tokenize()` to the paragraph stored in the variable `data`,
the output is a list of sentences. Each element in the list represents one sentence
from the original text.

Sentence tokenization is useful in many NLP tasks such as text summarization,
question answering, and machine translation, where processing is performed
at the sentence level rather than the paragraph level.

## Word Tokenization

Word tokenization splits text into individual words.


```python
words=word_tokenize(data)
print("No of Words in Corpus:",len(words))
print("Words in Corpus:\n",words)
```

```python
doc_to_words = [word_tokenize(doc) for doc in documents]
print("Words in Corpus grouped by sentence:")
for i in doc_to_words:
    print(i)
```


In this notebook we use the `word_tokenize()` function to convert the paragraph
into a list of word tokens. This tokenizer handles punctuation and contractions
better than simple string splitting.

We also apply `word_tokenize()` to each sentence separately to observe how the
paragraph structure is preserved when tokenization is performed sentence by sentence.

Word tokenization is a fundamental step in NLP because most text analysis tasks
such as text classification, sentiment analysis, and language modeling operate
at the word level.

## WordPunct Tokenizer

The `wordpunct_tokenize()` function separates words and punctuation into
different tokens.

Unlike `word_tokenize()`, punctuation marks such as commas, apostrophes,
and periods are treated as independent tokens.

This tokenizer is useful when punctuation itself carries meaning or when
a more detailed separation between words and symbols is required.

```python
words_punct=wordpunct_tokenize(data)
print("No of Words in Corpus:",len(words_punct))
print("Words in Corpus:\n",words_punct)
```

```python
doc_to_words_punct=[wordpunct_tokenize(doc) for doc in documents]

print("Words in Corpus grouped by sentence:")
for i in doc_to_words_punct:
    print(i)
```

## Treebank Word Tokenizer

The `TreebankWordTokenizer` follows the Penn Treebank tokenization rules,
which are widely used in NLP research datasets.

```python
treebank_tokenizer=TreebankWordTokenizer()
treebank_tokens=treebank_tokenizer.tokenize(data)

print("No of Words in Corpus:",len(treebank_tokens))
print("Words in Corpus:\n",treebank_tokens)
```

```python
doc_to_treebank_tokens=[treebank_tokenizer.tokenize(doc) for doc in documents]

print("Words in Corpus grouped by sentence:")
for i in doc_to_treebank_tokens:
    print(i)
```

Compared to `wordpunct_tokenize()`, it handles punctuation and contractions
in a more linguistically meaningful way.

In this notebook we apply this tokenizer to both the entire paragraph and
to individual sentences to observe how the tokenization differs from other methods.

## Regexp Tokenizer

`RegexpTokenizer` allows tokenization based on custom regular expressions.

In this notebook we use the regular expression `r'\w+'`, which extracts
only word characters and removes punctuation.

This tokenizer provides full control over how tokens are defined and is
particularly useful when preprocessing data for machine learning models
where punctuation may not be required.

```python
regexp_tokenizer = RegexpTokenizer(r'\w+')
regexp_tokens = regexp_tokenizer.tokenize(data)
print("No of Words in Corpus:",len(regexp_tokens))
print("Words in Corpus:\n",regexp_tokens)
```

```python
doc_to_regexp_tokens=[regexp_tokenizer.tokenize(doc) for doc in documents]

print("Words in Corpus grouped by sentence:")
for i in doc_to_regexp_tokens:
    print(i)
```

## Tweet Tokenizer

The `TweetTokenizer` from NLTK is designed to handle text commonly found on social media platforms such as Twitter.
Unlike standard tokenizers, it can properly process hashtags, user mentions, emojis, and repeated characters that
often appear in informal online communication.

In this notebook, the `TweetTokenizer` is applied to the sample text stored in the variable `tweet_data` to demonstrate
how it splits the text into tokens while preserving important elements such as punctuation and special characters.
Although the example text is not a tweet, this tokenizer shows how NLTK provides specialized tools for processing
noisy and unstructured text that appears in social media.

This tokenizer is commonly used in tasks such as sentiment analysis, social media monitoring, and text classification
where maintaining the structure of informal text is important.

```python
tweet_data="""The TweetTokenizer from NLTK is an indispensable tool in Natural Language Processing (NLP) specifically designed to handle
the noisy and informal nature of social media text 🐦. Unlike generic tokenizers that break up hashtags or discard emojis, this
rule-based tokenizer intelligently preserves hashtags (e.g., #NLP), user mentions (@user), and handles emoticons (e.g., :-), ✨) as
single, meaningful units 💡. It is particularly effective for sentiment analysis tasks because it maintains the emotional context of
emojis, which are often lost with standard word tokenization 💔➡😢. Furthermore, the TweetTokenizer offers flexible options, such as
reducing the length of excessively repeated characters (e.g., "coooooool" becomes "coool") and converting text to lowercase to
standardize data 🧠. It effortlessly manages punctuation and complex strings, breaking down long, unstructured posts into
clean, actionable tokens 🏗️. Whether you're working on sentiment classification, trend analysis, or building a chat bot, this tool
is the ultimate choice for cleaning up raw Twitter data and making it ready for modeling 🚀✅."""
```

```python
tweet_tokenizer=TweetTokenizer()
tweet_tokens=tweet_tokenizer.tokenize(tweet_data)
print("No of Words in Corpus:",len(tweet_tokens))
print("Words in Corpus:\n",tweet_tokens)
```

```python
docs_=sent_tokenize(tweet_data)
print("No of Sentences in Corpus:",len(docs_))
print("Sentences in Corpus:")
for doc in docs_:
    print(doc)
```

```python
doc_to_tweet_tokens=[tweet_tokenizer.tokenize(doc) for doc in docs_]

print("Words in Corpus grouped by sentence:")
for i in doc_to_tweet_tokens:
    print(i)
```

## Toktok Tokenizer

`ToktokTokenizer` is a simple and efficient tokenizer that is commonly
used in NLP pipelines.

It separates punctuation from words but keeps tokens relatively clean
and easy to process.

Compared to other tokenizers, it is faster and produces consistent results,
making it useful when working with large text datasets.

```python
toktok_tokenizer=ToktokTokenizer()
toktok_tokens=toktok_tokenizer.tokenize(data)
print("No of Words in Corpus:",len(toktok_tokens))
print("Words in Corpus:\n",toktok_tokens)
```

```python
doc_to_toktok_tokens=[toktok_tokenizer.tokenize(doc) for doc in documents]

print("Words in Corpus grouped by sentence:")
for i in doc_to_toktok_tokens:
    print(i)
```

## Whitespace Tokenizer

`WhitespaceTokenizer` splits text based only on whitespace characters.

Unlike other tokenizers, it does not consider punctuation or linguistic rules.
As a result, punctuation marks remain attached to words.

This tokenizer is extremely simple and fast, but it is usually not suitable
for complex NLP tasks where proper word separation is important.

```python
whitespace_tokenizer=WhitespaceTokenizer()
whitespace_tokens=whitespace_tokenizer.tokenize(data)
print("No of Words in Corpus:",len(whitespace_tokens))
print("Words in Corpus:\n",whitespace_tokens)
```

```python
doc_to_white_tokens=[whitespace_tokenizer.tokenize(doc) for doc in documents]

print("Words in Corpus grouped by sentence:")
for i in doc_to_white_tokens:
    print(i)
```

## Punkt Sentence Tokenizer

`PunktSentenceTokenizer` is an unsupervised algorithm for detecting
sentence boundaries.

It learns sentence segmentation rules from text and is commonly used
for sentence tokenization in many NLP applications.

In this notebook we create a tokenizer object and apply it directly
to the paragraph to observe how it identifies sentence boundaries.

```python
punktsent_tokenizer = PunktSentenceTokenizer()

punkt_sentences = punktsent_tokenizer.tokenize(data)

for doc in punkt_sentences:
    print(doc)
```

```python
def print_tokens(name, tokens):
    print("\n", name)
    print("Number of tokens:", len(tokens))
    print(tokens)
```

```python
tokenizers = {
    "TreebankWordTokenizer": TreebankWordTokenizer(),
    "RegexpTokenizer": RegexpTokenizer(r'\w+'),
    "TweetTokenizer": TweetTokenizer(),
    "ToktokTokenizer": ToktokTokenizer(),
    "WhitespaceTokenizer": WhitespaceTokenizer(),
}

for name, tokenizer in tokenizers.items():
    tokens = tokenizer.tokenize(data)
    print_tokens(name, tokens)
```

```python
tokenizers = {
    "PunktTokenizer": PunktTokenizer(),
    "PunktSentenceTokenizer": PunktSentenceTokenizer(),
}

for name, tokenizer in tokenizers.items():
    tokens = tokenizer.tokenize(data)
    print(name)
    for doc in tokens:
        print(doc)
    print("\n")
```

## Blankline Tokenizer

`BlanklineTokenizer` splits text into blocks whenever blank lines are encountered.

To demonstrate this tokenizer, we created a second text variable that contains
multiple blank lines between paragraphs.

Each block of text separated by blank lines becomes a separate token.
This tokenizer is useful when processing structured documents where
paragraph separation is important.

```python
blankline_data="""Personality development encompasses the dynamic construction and deconstruction of integrative characteristics that distinguish an individual in terms of interpersonal behavioral traits.
        Personality development is ever-changing and subject to contextual factors and life-altering experiences.


        Personality development is also dimensional in description and subjective in nature.

        That is, personality development can be seen as a continuum



        varying in degrees of intensity and change.








        It is subjective in nature because its conceptualization is rooted in social norms of expected behavior, self-expression, and personal growth.



        The dominant viewpoint in personality psychology indicates that personality emerges early and continues to develop across one's lifespan.
        Adult personality traits are


        believed to have a basis in infant temperament, meaning that individual differences in disposition and behavior appear early in life, potentially before language of conscious self-representation develop.
        The Five Factor Model of personality maps onto the dimensions of childhood temperament.
        This suggests that individual differences in levels of the corresponding personality traits (neuroticism, extraversion, openness to experience, agreeableness, and conscientiousness) are present from young ages."""
```

```python
tokenizer=BlanklineTokenizer()
blank_tokens=tokenizer.tokenize(blankline_data)
for doc in blank_tokens:
    print(doc)
```
