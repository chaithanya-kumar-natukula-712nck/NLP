```python
from nltk.stem import PorterStemmer,SnowballStemmer,LancasterStemmer,RegexpStemmer,LancasterStemmer
```

```python
words = [
"playing","played","player",
"studies","studying","study",
"happiness","happy",
"running","runner",
"universities","university",
"analysis","analyze",
"quickly","quickness",
"maximum","maximize"
]
data=['playing''played','player','studies','studying','happiness','quickly','runner','universities']
```

```python
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()
porter_stemmer_words_1 = [porter_stemmer.stem(word) for word in words]
print("Original words:", words)
print("Stemmed words:", porter_stemmer_words_1)

porter_stemmer_words_2 = [porter_stemmer.stem(data_) for data_ in data]
snowball_stemmer = SnowballStemmer(language='english')

print("Original words:", data)
print("Stemmed words:", porter_stemmer_words_2)
```

```python
snowball_stemmer = SnowballStemmer(language='english')
snowball_stemmer_words_1 = [snowball_stemmer.stem(word) for word in words]

print("Original words:", words)
print("Stemmed words:", snowball_stemmer_words_1)

snowball_stemmer_words_2 = [snowball_stemmer.stem(data_) for data_ in data]
print("\n")

print("Original words:", data)
print("Stemmed words:", snowball_stemmer_words_2)
```

```python
rule = r'ing$'
regexp_stemmer = RegexpStemmer(rule)
regexp_stemmer_words_1 = [regexp_stemmer.stem(word) for word in words]

print("Original words:", words)
print("Stemmed words:", regexp_stemmer_words_1)

regexp_stemmer_words_2 = [regexp_stemmer.stem(data_) for data_ in data]
print("\n")

print("Original words:", data)
print("Stemmed words:", regexp_stemmer_words_2)
```

```python
from nltk.stem import LancasterStemmer

lancaster_stemmer = LancasterStemmer()
stemmed_words_1 = [lancaster_stemmer.stem(word) for word in words]

print("Original words:", words)
print("Stemmed words:", stemmed_words_1)

stemmed_words_2 = [lancaster_stemmer.stem(data_) for data_ in data]
print("\n")
print("Original words:", data)
print("Stemmed words:", stemmed_words_2)
```

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")


print("Word \tPorter \tSnowball \tLancaster")

for word in word_list:
    print(word,"\t",
          porter.stem(word),"\t",
          snowball.stem(word),"\t",
          lancaster.stem(word))
    print("\n")
```
